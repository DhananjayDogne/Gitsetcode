```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB'] # x  represent company
revenue=[90,136,89,27]  #y represent revenue
profit=[40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(company, revenue)
```

```python

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")


plt.bar(xpos,revenue, label="Revenue")

plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(xpos-0.2,revenue,  label="Revenue")
plt.bar(xpos+0.2,profit, label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Techn Stocks')
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4, label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Techn Stocks')
plt.legend()
```

```python
plt.barh(xpos-0.2,revenue,  label="Revenue")
plt.barh(xpos+0.2,profit,  label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Techn Stocks')
plt.legend()
```

```python

plt.xticks(xpos,company)
plt.title('US Technology Stocks')
plt.ylabel("Revenue(Bln)")
plt.xlabel("Company(Bln)")

plt.barh(xpos,company, label="Revenue")
plt.barh(xpos,revenue, label="profit")


plt.legend()
```

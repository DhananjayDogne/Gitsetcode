```python
# DRAW A PIE CHART AS PERFECT CIRCLE


exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

# perfect circle
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.show
```

```python
# show percentages for every pie. Specify radius to increase chart size

plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.1f%%')
plt.show()
```

```python
# EXPLODE_specifiy no, of value

plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.1f%%', explode=[0,0.2,0.1,0,0])
plt.show()
```

```python
# counter clock and angle property
```

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0], startangle=45)
plt.show()
```

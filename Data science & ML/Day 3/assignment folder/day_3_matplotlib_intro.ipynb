```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=5, linestyle="dotted")
```

```python
plt.xlabel('Days')
plt.ylabel("temperature")
plt.title('weather')
plt.plot(x,y, color='yellow',linewidth=5, linestyle="dotted")
```

```python
plt.plot(x,y, 'r*--')
```

```python
plt.plot(x,y, 'g+--')
```

```python
plt.plot(x,y, color='#008926', marker='D', linestyle='')
```

```python
plt.plot(x,y, color='#008926', marker='+', linestyle='',markersize=30)
```

```python
# aplpha controls the transparency

plt.plot(x,y, color='green', alpha=0.2 )
```
